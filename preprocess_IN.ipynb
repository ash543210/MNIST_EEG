```python
import pandas as pd 
dataset = pd.read_csv("documents/IN.txt", sep="\t") 
top_row = dataset.columns
print(dataset)
```

```python
labels = ['id', 'event', 'device', 'channel', 'code', 'size', 'data']
label_change = {}
```

```python
def change_columns():
    i = 0
    for key in labels:
        label_change[top_row[i]] = key
        i += 1
    dataset.rename(columns = label_change, inplace = True)
```

```python
change_columns()
```

```python
def add_top_row():
    data = {}
    i = 0
    for key in labels:
        data[key] = top_row[i]
        i += 1
    new_row = pd.DataFrame(data, index=[0])
    return new_row
```

```python
dataset = pd.concat([add_top_row() , dataset.loc[:]]).reset_index(drop=True)
```

```python
dataset = dataset.astype({'size' : 'int64', 'data' : 'string', 'channel' : 'string', 'code' : 'string', 'id' : 'string', 'event': 'string', 'device' : 'string'})
```
